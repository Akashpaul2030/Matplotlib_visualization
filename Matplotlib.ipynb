```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.arange(1,10)
```

```python
y = np.array(x**2)
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.scatter(x,y)
```

```python
plt.hist(x)
```

```python
plt.hist(y)
```

```python
plt.plot(x,y)
plt.xlabel("X_axis")
plt.ylabel("Y_axis")
plt.title("Plotting x,y")
```

```python
plt.plot(x,x**2,label="x**2")
plt.plot(x,x**3,label="x**3")
plt.xlabel("X_axis")
plt.ylabel("Y_axis")
plt.title("Plotting x,y")
plt.legend()
```

```python
plt.subplot(1,2,1)
plt.plot(x,x**2)
plt.xlabel("X_axis")
plt.ylabel("x**2_axis")
plt.title("Plotting x,x**2")

plt.subplot(1,2,2)
plt.plot(x,x**3)
plt.xlabel("X_axis")
plt.ylabel("x**2_axis")
plt.title("Plotting x,x**2")
```

```python
plt.subplot(2,2,1)
plt.plot(x,x**2)
plt.xlabel("X_axis")
plt.ylabel("x**2_axis")
plt.title("Plotting x,x**2")

plt.subplot(2,2,2)
plt.plot(x,x**3)
plt.xlabel("X_axis")
plt.ylabel("x**3_axis")
plt.title("Plotting x,x**3")

plt.subplot(2,2,3)
plt.plot(x,x**4)
plt.xlabel("X_axis")
plt.ylabel("x**4_axis")
plt.title("Plotting x,x**4")

plt.subplot(2,2,4)
plt.plot(x,x**5)
plt.xlabel("X_axis")
plt.ylabel("x**5_axis")
plt.title("Plotting x,x**5")

plt.tight_layout()
```
