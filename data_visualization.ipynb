1. Bar Chart - Total Sales:
Given a list of products and their corresponding sales quantities, create a bar chart to visualize the total sales of each product.
INPUT
products = ["Product A", "Product B", "Product C", "Product D"]
sales = [150, 200, 80, 120]
Output: (Bar chart displaying total sales for each product)

```python
import matplotlib.pyplot as plt

# Create a list of products
products = ["Product A", "Product B", "Product C", "Product D"]

# Create a list of sales quantities
sales = [150, 200, 80, 120]

# Create a bar chart
plt.bar(products, sales)

# Add a title to the chart
plt.title("Total Sales")

# Add labels to the axes
plt.xlabel("Product")
plt.ylabel("Sales Quantity")

# Show the chart
plt.show()
```

2. Histogram - Exam Scores:
Given a list of exam scores (integer values), create a histogram to visualize the distribution of scores.
INPUT
scores = [85, 90, 78, 92, 88, 85, 78, 95, 84, 88, 90, 88, 92, 85]
Output: (Histogram showing the frequency of scores in different bins)

3. Line Plot - Stock Prices:
Given a list of dates and corresponding stock prices, create a line plot to visualize the trend of stock prices over time.
Input:
dates = ["2023-07-01", "2023-07-02", "2023-07-03", ...]
stock_prices = [100.2, 102.5, 105.1, ...]
Output: (Line plot showing the trend of stock prices over time)


```python
%matplotlib inline

# Input data
dates = ["2023-07-01", "2023-07-02", "2023-07-03"]
stock_prices = [100.2, 102.5, 105.1]

# Create a line plot
plt.plot(range(len(dates)), stock_prices, marker='o', color="#00FF00")

# Set the title and axis labels
plt.title("Stock Prices")
plt.xlabel("Date")
plt.ylabel("Price")

# Set the x-axis ticks and labels
plt.xticks(range(len(dates)), dates)

# Rotate the x-axis date labels for better readability (optional)
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

5. Pie Chart - Sales Proportions:
Given a dictionary containing the sales proportions of different products, create a pie chart to visualize the proportions.
Input:
sales_proportions = {"Product A": 0.35, "Product B": 0.25, "Product C": 0.15, "Product D": 0.25}
Output: (Pie chart showing the proportions of sales for each product)


```python
# Use this magic command for inline plotting in Google Colab
%matplotlib inline

# Input data
sales_proportions = {"Product A": 0.35, "Product B": 0.25, "Product C": 0.15, "Product D": 0.25}

# Extract product names and proportions from the dictionary
products = list(sales_proportions.keys())
proportions = list(sales_proportions.values())

# Create a pie chart with a smaller size
plt.figure(figsize=(6, 6))  # Adjust the figsize values as per your preference

# Customize the colors of the pie chart slices
colors = ["#ff9999", "#66b3ff", "#99ff99", "#ffcc99"]

# Create the pie chart
plt.pie(proportions, labels=products, autopct='%1.1f%%', colors=colors)

# Set the title
plt.title("Sales Proportions of Different Products")

# Show the plot
plt.show()
```

6. Heatmap from Correlation Matrix:
Write a Python function that takes a 2D list representing a correlation matrix and generates a heatmap to visualize the correlation values.
Example:
Input:
correlation_matrix = [
    [1.0, 0.8, 0.6, 0.2],
    [0.8, 1.0, 0.4, 0.3],
    [0.6, 0.4, 1.0, 0.7],
    [0.2, 0.3, 0.7, 1.0]
]
Output:
(Display a heatmap representing the correlation values)


```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_correlation_heatmap(correlation_matrix):
    plt.figure(figsize=(8, 6))  # Set the size of the heatmap

    # Create the heatmap using seaborn
    sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", annot_kws={"size": 10},
                xticklabels=False, yticklabels=False)

    # Set the title
    plt.title("Correlation Matrix Heatmap")

    # Show the plot
    plt.show()
```


7. Box Plot with Outlier Detection:
Write a Python function that takes a list of numerical data and generates a box plot to visualize the distribution, along with identifying and marking the outliers.
Example:
Input:
data = [10, 15, 18, 20, 22, 25, 28, 30, 100, 200]
Output:
(Display a box plot showing the distribution of data and markingtheoutliers)

```python
correlation_matrix = [
    [1.0, 0.8, 0.6, 0.2],
    [0.8, 1.0, 0.4, 0.3],
    [0.6, 0.4, 1.0, 0.7],
    [0.2, 0.3, 0.7, 1.0]
]

plot_correlation_heatmap(correlation_matrix)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def box_plot_with_outlier_detection(data):


    # Calculate the interquartile range (IQR)
    iqr = np.percentile(data, 75) - np.percentile(data, 25)

    # Calculate the lower and upper bounds of the whiskers
    lower_bound = np.percentile(data, 25) - 1.5 * iqr
    upper_bound = np.percentile(data, 75) + 1.5 * iqr

    # Identify the outliers
    outliers = [x for x in data if x < lower_bound or x > upper_bound]

    # Create a box plot
    plt.boxplot(data)

    # Mark the outliers
    plt.plot(np.ones(len(outliers)), outliers, "ro", label="Outliers")

    # Set the title and labels
    plt.title("Box Plot with Outlier Detection")
    plt.xlabel("Data")
    plt.ylabel("Values")

    # Show the legend for outliers
    if len(outliers) > 0:
        plt.legend()

    # Show the plot
    plt.show()


if __name__ == "__main__":
    # Example data
    data = [10, 15, 18, 20, 22, 25, 28, 30, 100, 200]

    # Plot the box plot with outlier detection
    box_plot_with_outlier_detection(data)
```
